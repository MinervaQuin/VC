```python
from deepface import DeepFace
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
def verifyFaces(img1_p, img2_p):

    result = DeepFace.verify(img1_p, img2_p)
    print("result: ", result)
    verification = list(result.values())[0]


    if verification:
        print("the are the same person")
    else:
        print("they are different persons")


img1_p = "./images/boy-1262810_1280.jpg"
img2_p = "./images/boy-1262811_1280.jpg"

verifyFaces(img1_p, img2_p)



    
```

```python
def search_face(img1_p, directory):
    metrics = ["cosine", "euclidean", "euclidean_l2"]
    print(directory)
    #face recognition
    df = DeepFace.find(img_path = img1_p, 
            db_path = directory, 
            distance_metric = metrics[2],
            enforce_detection=False
    )

    img_paths = df[0]['identity']
    img_url = []

    for url in img_paths:
        # print(url)
        img_url.append(url)
        
    return img_url


search_face("./images/boy-1262810_1280.jpg", "./images/")
```

```python
def elementos_repetidos(arr1, arr2):
    set_arr1 = set(arr1)
    set_arr2 = set(arr2)
    elementos_comunes = set_arr1.intersection(set_arr2)
    return list(elementos_comunes)
```

```python
def search_two_faces(img1_p, img2_p, directory):
    url_im1 = search_face(img1_p, directory)
    url_im2 = search_face(img2_p, directory)
    resultado = elementos_repetidos(url_im1, url_im2)
    return resultado


```

```python
from tkinter import Tk, Button, Label, filedialog, Frame
from PIL import Image, ImageTk
import os


class InterfazGrafica:

    def __init__(self, ventana):
        self.ventana = ventana
        self.ventana.title("Seleccionar Directorio/Imagen")

        self.directorio_path = ""
        self.photo_path = ""

        # Botones
        self.btn_seleccionar_directorio = Button(ventana, text="Seleccionar Directorio", command=self.seleccionar_directorio)
        self.btn_seleccionar_directorio.pack(pady=10)
        self.etiqueta_directorio = Label(ventana, text="Directorio seleccionado:")
        self.etiqueta_directorio.pack(pady=5)

        self.btn_seleccionar_imagen = Button(ventana, text="Seleccionar Imagen", command=self.seleccionar_imagen)
        self.btn_seleccionar_imagen.pack(pady=10)
        self.etiqueta_imagen = Label(ventana, text="Imagen seleccionada:")
        self.etiqueta_imagen.pack(pady=5)

        self.btn_buscar_persona =  Button(ventana, text="Buscar", command=self.buscar_persona)
        self.btn_buscar_persona.pack(pady=10)
        self.label = Label(self.ventana, text="Imagen encontrada: ")


        # Crear un Frame para contener los labels
        self.frame_labels = Frame(ventana)
        self.frame_labels.pack(pady=10)


        ancho_ventana = 800
        alto_ventana = 500
        x_pos = (ventana.winfo_screenwidth() - ancho_ventana) // 2
        y_pos = (ventana.winfo_screenheight() - alto_ventana) // 2
        ventana.geometry(f"{ancho_ventana}x{alto_ventana}+{x_pos}+{y_pos}")

        # Establecer ancho y alto mínimo
        ventana.minsize(300, 150)


    def seleccionar_directorio(self):
        directorio = filedialog.askdirectory(title="Seleccionar directorio de imágenes")
        
        if directorio:
            self.directorio_path=directorio
            self.etiqueta_directorio.config(text=f"Directorio seleccionado: {directorio}")
            # Puedes llamar a tu función de extraer directorio aquí

    def seleccionar_imagen(self):
        archivo_imagen = filedialog.askopenfilename(
            title="Seleccionar archivo de imagen",
            filetypes=[("Archivos de imagen", "*.jpg;*.jpeg;*.png")]
        )
        if archivo_imagen:
            self.photo_path=archivo_imagen
            self.etiqueta_imagen.config(text=f"Imagen seleccionada: {archivo_imagen}")
            # Puedes llamar a tu función de extraer imagen aquí

    def buscar_persona(self):

        if not self.photo_path or not self.directorio_path:
            return 
        
        for widget in self.frame_labels.winfo_children():
            widget.destroy()

        img_paths = search_face(self.photo_path,self.directorio_path)

        for path in img_paths:
            label = Label(self.frame_labels, text=f"Imagen encontrada: {path}")
            label.pack(pady=5)
            label.bind("<Button-1>", lambda event, path=path: self.abrir_imagen_label(event, path))
            label.bind("<Enter>", lambda event, label=label: self.cambiar_color_label(label, "blue", True))
            label.bind("<Leave>", lambda event, label=label: self.cambiar_color_label(label, "black", False))

    def abrir_imagen_label(self, event, img_path):
        if os.path.exists(img_path):
            img = Image.open(img_path)
            img.show()

    def cambiar_color_label(self, label, color, subrayar):
        label.config(fg=color, underline=subrayar)
        
```
